```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
pce_data = pd.read_csv('CPI.csv')

# Extract date part and convert using format for abbreviated month
date_str = pce_data['Release Date'].str.extract(r'([A-Za-z]{3} \d{1,2}, \d{4})')[0]
pce_data['Parsed Date'] = pd.to_datetime(date_str, format='%b %d, %Y')

# Sort by parsed date
pce_data = pce_data.sort_values('Parsed Date')

# Create Year-Month column for x-axis
pce_data['YearMonth'] = pce_data['Parsed Date'].dt.strftime('%Y-%m')

plt.figure(figsize=(14, 5))

plt.plot(pce_data['YearMonth'], pce_data['Actual'], label='Actual', marker='o')
plt.plot(pce_data['YearMonth'], pce_data['Forecast'], label='Forecast', marker='x')

plt.title('PCE: Actual vs Forecast Over Time')
plt.xlabel('Year-Month')
plt.ylabel('PCE Values')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.xticks(rotation=90, ha='right')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
pce_data = pd.read_csv('CPI.csv')

# Extract date part and convert to datetime
date_str = pce_data['Release Date'].str.extract(r'([A-Za-z]{3} \d{1,2}, \d{4})')[0]
pce_data['Parsed Date'] = pd.to_datetime(date_str, format='%b %d, %Y')

# Extract month and year
pce_data['Month'] = pce_data['Parsed Date'].dt.month
pce_data['Year'] = pce_data['Parsed Date'].dt.year

# Create a string like "2018 Jan"
pce_data['YearMonthStr'] = pce_data['Year'].astype(str) + ' ' + pce_data['Parsed Date'].dt.strftime('%b')

# Sort data by Month first, then Year
pce_data = pce_data.sort_values(by=['Month', 'Year'])

# Plot
plt.figure(figsize=(16, 6))

plt.plot(pce_data['YearMonthStr'], pce_data['Actual'], marker='o', label='Actual')
plt.plot(pce_data['YearMonthStr'], pce_data['Forecast'], marker='x', label='Forecast')

plt.title('PCE: Actual vs Forecast (Grouped by Month, then Year)')
plt.xlabel('Year Month')
plt.ylabel('PCE Values')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```
