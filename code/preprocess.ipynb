```python
def rename_and_merge(base_df, df_to_merge, prefix):
    # Rename columns except 'datetime'
    renamed_df = df_to_merge.rename(columns={col: f'{prefix}_{col}' for col in df_to_merge.columns if col != 'datetime'})
    # Merge with base_df on datetime
    merged = base_df.merge(renamed_df, on='datetime', how='left')
    return merged

# Ensure all datetime columns are datetime type
mergedCPI['datetime'] = pd.to_datetime(mergedCPI['datetime'])
GDP['datetime'] = pd.to_datetime(GDP['datetime'])
NFP['datetime'] = pd.to_datetime(NFP['datetime'])
Interest_Rate['datetime'] = pd.to_datetime(Interest_Rate['datetime'])
PPI['datetime'] = pd.to_datetime(PPI['datetime'])
PCE['datetime'] = pd.to_datetime(PCE['datetime'])

# Merge each economic dataframe into mergedCPI
mergedCPI = rename_and_merge(mergedCPI, GDP, 'GDP')
mergedCPI = rename_and_merge(mergedCPI, NFP, 'NFP')
mergedCPI = rename_and_merge(mergedCPI, Interest_Rate, 'InterestRate')
mergedCPI = rename_and_merge(mergedCPI, PPI, 'PPI')
mergedCPI = rename_and_merge(mergedCPI, PCE, 'PCE')

# View merged dataframe
mergedCPI.head()
```

```python
def check_availability(df, price_data, name):
    df['in_Price_data'] = df['datetime'].isin(price_data['datetime'])
    available_count = df['in_Price_data'].sum()
    total_count = len(df)
    print(f"{name} records available in Price_data: {available_count} out of {total_count}")

# Assuming Price_data['datetime'] is already datetime type
Price_data['datetime'] = pd.to_datetime(Price_data['datetime'])

# Check for each dataset
check_availability(CPI, Price_data, 'CPI')
check_availability(GDP, Price_data, 'GDP')
check_availability(NFP, Price_data, 'NFP')
check_availability(Interest_Rate, Price_data, 'Interest Rate')
check_availability(PPI, Price_data, 'PPI')
check_availability(PCE, Price_data, 'PCE')
```

```python
# Function to get and print unavailable records
def print_unavailable(df, name):
    unavailable = df[~df['in_Price_data']]
    print(f"\n{name} records NOT available in Price_data:")
    print(unavailable)
    return unavailable

# Example for CPI
unavailable_cpi = print_unavailable(CPI, 'CPI')

# Similarly for other datasets
unavailable_gdp = print_unavailable(GDP, 'GDP')
unavailable_nfp = print_unavailable(NFP, 'NFP')
unavailable_interest = print_unavailable(Interest_Rate, 'Interest Rate')
unavailable_ppi = print_unavailable(PPI, 'PPI')
unavailable_pce = print_unavailable(PCE, 'PCE')
```
